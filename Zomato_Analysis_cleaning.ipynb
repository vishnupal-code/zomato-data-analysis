```python
# --- extract a standardized primary 10-digit phone number (with or without a '+91' prefix) from potentially messy string data
# in the original 'phone' column and store the extracted number (or 'Not Available' if no valid number is found) in a new column
# called 'clean_phone'---
```

```python
import re

def extract_primary_phone(text):
    # Step 1: Convert to string (in case of NaN)
    text = str(text)

    # Step 2: Remove newline, carriage returns and extra spaces
    text = text.replace('\r', ' ').replace('\n', ' ').strip()

    # Step 3: Regular expression to match:
    #   a) +91 followed by optional space and 10 digits
    #   b) OR any 10-digit number
    match = re.search(r'(\+91\s*\d{10}|\d{10})', text)

    # Step 4: Return the matched number or 'Not Available'
    return match.group(0) if match else 'Not Available'

# Apply this function to your column and create a new cleaned one
df['clean_phone'] = df['phone'].apply(extract_primary_phone)
```
